```python
import oikos as ok
```

```python
# Crear lienzo
canvas = ok.Canvas()

# Configurar
canvas.setLabels(xlabel="Cantidad (Q)", ylabel="Precio (P)", title="Equilibrio de Mercado")
canvas.setRange(x_range=(0, 150), y_range=(0, 150))
canvas.setTicks(x_step=10, y_step=10)  # Saltos de 10 en 10

# Añadir funciones (¡se grafican automáticamente!)
demanda = ok.Demand(r"Q = 120 - P")
oferta = ok.Supply(r"Q = 2P")

canvas.add(demanda, label="Demanda")
canvas.add(oferta, label="Oferta")

# Añadir punto de equilibrio
canvas.addPoint(80, 40, label="Equilibrio", color="black", size=12)

# Mostrar
canvas.plot()
```
